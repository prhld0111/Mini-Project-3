```python
ipl_data = pd.read_csv("C:\\Users\\prhld\\Downloads\\IPL_Dataset.csv")
ipl_data.head()
```

```python
max_matches = ipl_data['Mat'].max()
print(f"Q1. Maximum number of matches played by an individual player in a season: {max_matches}")
```

```python
highest_average = ipl_data['Avg'].max()
print(f"Q2. Highest average of a player in a season: {highest_average}")
```

```python
average_matches = ipl_data['Mat'].mean()
print(f"Q3. Average number of matches played by players in a season: {average_matches}")
```

```python
correlation_matrix = ipl_data.corr()
print(f"Q4. Correlation between all features:\n{correlation_matrix}")
```

```python
filtered_players_q5 = ipl_data[(ipl_data['Avg'] > 50) & (ipl_data['SR'] > 120)]
print(f"Q5. Players with average > 50 and strike rate > 120:\n{filtered_players_q5}")
```

```python
filtered_players_q6 = ipl_data[(ipl_data['Avg'] > 40) & (ipl_data['BF'] > 100)]
print(f"Q6. Players with average > 40 and balls faced > 100:\n{filtered_players_q6}")
```

```python
filtered_players_q7 = ipl_data[(ipl_data['Runs'] > 300) & (ipl_data['Avg'] > 45) & (ipl_data['SR'] > 130)]
print(f"Q7. Players with more than 300 runs, average > 45, and strike rate > 130:\n{filtered_players_q7}")
```

```python
century_scorers = ipl_data[ipl_data['100'] >= 1]
print(f"Q8. Players who scored at least one century:\n{century_scorers}")
```

```python
half_century_scorers = ipl_data[ipl_data['50'] >= 4]
print(f"Q9. Players who scored at least 4 half centuries:\n{half_century_scorers}")
```

```python
filtered_players_q10 = ipl_data[(ipl_data['4s'] > 45) & (ipl_data['6s'] > 10)]
print(f"Q10. Players with more than 45 boundaries and more than 10 sixes:\n{filtered_players_q10}")
```

```python

plt.hist(ipl_data['Mat'], bins=20, edgecolor= 'red')
plt.title('Histogram of Number of Matches Played in a Season by Players')
plt.xlabel('Number of Matches')
plt.ylabel('Count')
```

```python

plt.hist(ipl_data['BF'], bins=20, edgecolor='green')
plt.title('Histogram of Balls Faced by Players')
plt.xlabel('Number of Balls Faced')
plt.ylabel('Count')

```

```python
top_20_run_scorers = ipl_data.nlargest(20, 'Runs')
print(f"Q13. Top 20 players with most runs in a season:\n{top_20_run_scorers[['Player', 'Runs']]}")
```

```python
top_20_strike_rate = ipl_data.nlargest(20, 'SR')
plt.figure(figsize=(12, 6))
plt.bar(top_20_strike_rate['Player'], top_20_strike_rate['SR'])
plt.title('Top 20 Players with Best Strike Rate in a Season')
plt.xlabel('Player')
plt.ylabel('SR')
plt.xticks(rotation=45)
plt.show()
```

```python
top_30_six_hitters = ipl_data.nlargest(30, '6s')
print(f"Q15. Top 30 players with highest number of sixes:\n{top_30_six_hitters[['Player', '6s']]}")
```

```python

plt.scatter(ipl_data['Runs'],ipl_data['BF'], alpha=0.5, edgecolor='red')
plt.title('Scatter Plot of Runs Scored vs. Balls Faced in a Season')
plt.xlabel('Runs')
plt.ylabel('BF')
plt.show()
```

```python
import seaborn as sns

sns.kdeplot(ipl_data['Avg'], fill=True, edgecolor='red')
plt.title('KDE Plot for Averages in a Season')
plt.xlabel('Avg')
plt.ylabel('Density')
plt.show()
```
